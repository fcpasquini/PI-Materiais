## KNN

Alterar a função `vizinho_mais_proximo_()` para identificar se a variável resposta é contínua ou não, fazendo um KNN Classificador ou KNN Regressor de acordo com isso.

Para o KNN Classificador, retornar a probabilidade de pertencer a uma determinada classe. Já para o KNN Regressor, trazer uma tupla com a média ou mediana, de acordo com o parâmetro `est`, e o desvio padrão.

__Passo 0:__ Importando as bibliotecas necessárias.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
```

__Passo 1:__ Criando a função de identificação do KNN.

```python
def vizinho_mais_proximo_(d_treino, nova_instancia, var_resposta, k = 1, est = 'media'):
    '''
    Dado um conjunto de dados chamado d_treino, calcula a predição da classe para a nova_instancia.
    '''
    
    lista_distancia = []
    
    for lin in range(d_treino.shape[0]):
        
        d = 0

        for col in range(4):
            d += (d_treino.iloc[lin][col] - nova_instancia.iloc[col])**2

        d = np.sqrt(d)
        
        lista_distancia.append((d, d_treino.iloc[lin][var_resposta]))
    
    #lista_distancia = sorted(lista_distancia)
    #lista_distancia = lista_distancia[:k]
    #moda = mode([i[1] for i in lista_distancia]).mode
    
    return lista_distancia
```

__Passo 2:__ Importando e preparando o dataset Iris para utilização.

```python
df = pd.read_csv('iris.csv')
```

```python
kf = KFold(n_splits=2, random_state = 1, shuffle = True)
```

```python
x, y = kf.split(df)
```

```python
x_teste = df.drop(labels = x[0])
x_treino = df.drop(labels = x[1], inplace = False)
y_teste = df.drop(labels = y[0], inplace = False)
y_treino = df.drop(labels = y[1], inplace = False)
```

```python
lista_especies = ['setosa', 'virginica', 'versicolor']

contador = 1

for a in lista_especies:

    x_teste['species'] = x_teste['species'].replace(a, contador)
    x_treino['species'] = x_treino['species'].replace(a, contador)
    y_teste['species'] = y_teste['species'].replace(a, contador)
    y_treino['species'] = y_treino['species'].replace(a, contador)
    contador += 1
```

```python
vizinho_mais_proximo_(x_treino, x_teste, 'species', k = 1, est = 'media')
```
